```python
data_nur=data_nurbed.pivot_table(index=['district_name','variety'],values=['total_quantity'],aggfunc="sum")
data_nur=data_nur.reset_index()
```

```python
data_design=pd.read_csv(r"C:\\Users\\Edison New\Desktop\\HWG Q1 Reporting 2025\\Raw data sets from the app\design nurserybed summary.csv")
```

```python
data_design.columns
```

```python
data_design = data_design.melt(
       value_vars=['hot_pepper_tins_organic_pesticides', 'onions_bags_organic_pesticides', 'Spinach', 'AFRICAN SPIDER PLANT', 'Amaranths', 'Tomatoes', 'BLACK NIGHT SHADE', 'Purple Eggplants', 'Green Pepper', 'White Eggplants', 'Cabbage', 'Nakati', 'Sukuma Wiki'],
       value_name='total_quantity',
       var_name='variety',
       id_vars=['district_name', 'cohort', 'cycle']
)
```

```python
data_design_pivot=data_design.pivot_table(index=['district_name','variety'],values=['total_quantity'],aggfunc="sum")
```

```python
data_design_pivot = data_design_pivot.reset_index()
```

```python
data_design_pivot['variety'] = data_design_pivot['variety'].str.strip()
data_design_pivot.head(25)
```

```python
data_design_pivot['variety'].replace({'AFRICAN SPIDER PLANT':'African spider Plant',
                                  'BLACK NIGHT SHADE':'Black night shade'}, inplace=True)
```

```python
data_design_pivot = data_design_pivot[~data_design_pivot['variety'].isin(['hot_pepper_tins_organic_pesticides', 'onions_bags_organic_pesticides'])]
```

```python
data_design_pivot
```

```python
clean_data_nurserybed=data_design_pivot.merge(data_nur,on=['district_name','variety'],suffixes=('_design','_system'),how='outer')
```

```python
filepath = 'C:\\Users\\Edison New\\Desktop\\HWG Q1 Reporting 2025\\Final outputs\\nurserybedfinal analysis.csv'
clean_data_nurserybed.to_csv(filepath, index=False)
```
